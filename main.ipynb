## Crop frames

```python
import crop_frames

# ── Single file ──────────────────────────────────────────────────────────────
input_path = "../pirate-carrier-1.png"

frames_w = 8  # columns (frames per row)
frames_h = 8  # rows
frame_count = frames_w * frames_h

crop_frames.crop_frames(
    input_paths=input_path,    # str or list[str]
    frame_count=frame_count,
    frames_w=frames_w,
    frames_h=frames_h,
)
```

```python
import crop_frames

# ── Split spritesheet (foo-1.png, foo-2.png, ...) ────────────────────────────
# All files must share the same frame size.
# The union bounding box is computed across ALL frames from ALL files.
prefix = "../ballista-turret"
input_paths = [f"{prefix}-{i}.png" for i in range(1, 5)]   # -1.png, -2.png

frames_w = 10  # columns (frames per row)
frames_h = 16  # rows
frame_count = frames_w * frames_h

crop_frames.crop_frames(
    input_paths=input_paths,
    frame_count=frame_count,   # frames per file
    frames_w=frames_w,
    frames_h=frames_h,
)
```
