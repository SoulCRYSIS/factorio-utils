## Merge sprites

## Reduce frames

```python
import reduce_rotation

# Path to the spritesheet to reduce
input_path = "../graphics/entities/creatures/pirate-carrier/pirate-carrier-water-reflection.png"

# How many frames to skip?
# 4 means keep every 4th frame (e.g. 1, 5, 9... which is 0, 4, 8 in 0-indexed)
skip = 2

# Total frames in the input image (optional but helps detection)
total_frames = 32

reduce_rotation.reduce_rotation(
    input_path=input_path,
    skip=skip,
    frame_count=total_frames
)
```

## Shift frames

```python
import shift_frames

# Path to the spritesheet
input_path = "../graphics/entities/creatures/pirate-carrier/pirate-carrier-glow.png"

# Shift amount (cut first X frames and move to end)
shift = 8

# Total frames (optional if auto-detect works)
total_frames = 32

shift_frames.shift_frames(
    input_path=input_path,
    shift_amount=shift,
    frame_count=total_frames
)
```
